##Data Preparation

A quick peek at `county_facts` tells us that there are several problems with data: 

1.  There are rows with both county and state-specific information (i. e. `Alabama` and `Autaga County` which belongs to Alabama). Not to mention the very first row with `United States`. If we did nothing, state-specific information would be taken to be on par with county-specific information which would lead to significant outliers. 

 
2.   Related to the first, the second problem is that `state_abbreviation` has `NaN` values on rows with state-specific information. 

We will be exploiting the latter fact in order to filter out the state-specific information. 


```
county_facts.isna().any().head()
```

```
state_facts = county_facts[county_facts.isna().any(axis=1)]
```

```
state_facts.head()
```

```
county_facts = pd.concat([county_facts, state_facts]).drop_duplicates(keep=False)
```

```
county_facts.head()
```

```
state_facts = state_facts.drop(0).reset_index(drop=True).drop(columns=['fips', 'state_abbreviation'])
```

```
state_facts.head()
```

At this point we have separate datasets for county and state-specific information respectively: `county_facts` and `state_facts` 

Now, it is said that the keys to to winning the U.S. presidential election are the so-called swing states (for this reason also called battleground states). Based on Politico, the following were considered to be swing states for 2016:
*   Colorado
*   Florida
*   Iowa
*   Michigan
*   Nevada
*   New Hampshire
*   North Carolina
*   Ohio
*   Pennsylvania
*   Virginia
*   Wisconsin


```
swing_states = ['Colorado', 'Florida', 'Iowa', 'Michigan', 'Nevada', 'New Hampshire', 'North Carolina', 'Ohio', 'Pennsylvania', 'Virginia', 'Winconsin']
```

```
swing_state_facts = state_facts.loc[state_facts['area_name'].isin(swing_states)].reset_index(drop=True)
```

```
swing_state_facts
```

Our take on the project is to specifically target two specific features, namely, `High school graduate or higher`(%) and `Bachelor's degree or higher` (%) and to do so with respect to the voting results in swing states.

Let us further "zoom in" on the features we are interested in by making yet another dataset, `swing_state_educational_facts`:

```
swing_state_ed_facts = swing_state_facts[['area_name', 'PST045214', 'EDU635213', 'EDU685213']].copy().rename(columns={"PST045214": "population", "EDU635213": "high_school_percent", "EDU685213": "higher_ed_percent"})
```

```
swing_state_ed_facts
```

```
swing_state_ed_facts['high_school_grads'] = swing_state_ed_facts[['population', 'high_school_percent']].apply(
    lambda x: (x[0] * (x[1] / 100)), axis=1
).astype(int)

swing_state_ed_facts['high_school_dropouts'] = swing_state_ed_facts[['population', 'high_school_grads']].apply(
    lambda x: (x[0] - x[1]), axis=1
).astype(int)

swing_state_ed_facts['higher_school_graduates'] = swing_state_ed_facts[['population', 'higher_ed_percent']].apply(
    lambda x: (x[0] * (x[1] / 100)), axis=1
).astype(int)

swing_state_ed_facts['no_higher_school'] = swing_state_ed_facts[['population', 'higher_school_graduates']].apply(
    lambda x: (x[0] - x[1]), axis=1
).astype(int)
```

```
swing_state_ed_facts
```

## Data Visualization

Here we graph the ratio of high-school graduates to high-school dropouts in the swing states using `swing_state_ed_facts`:

## Feature Engineering

Let us now zoom in on the swing states within `primary results`.

```
swing_state_results = primary_results.loc[primary_results['state'].isin(swing_states)].reset_index(drop=True)
```

```
swing_state_results.head()
```

```
swing_state_results_numbers = swing_state_results.groupby(['state', 'party']).agg({'votes': 'sum'})
```

```
swing_state_results_numbers.head()
```

It seems that the Republican votes in Colorado has been left out for whatever reason. Let us group the `primary_results` by party so we can make sure that this has indeed to do with the data.

```
democrat_swing_results = swing_state_results[swing_state_results['party'] == 'Democrat'].groupby('state')['votes'].sum()
```

```
democrat_swing_results.head()
```

```
republican_swing_results = swing_state_results[swing_state_results['party'] == 'Republican'].groupby('state')['votes'].sum()
```

```
republican_swing_results.head()
```

As it seems to be indeed the case, we will unfortunately have to ignore the Colorado results thorough the remainder of the section:

```
swing_state_results = pd.concat([democrat_swing_results.rename('Democrat'), republican_swing_results.rename('Republican')], axis=1).drop('Colorado').rename_axis('area_name').reset_index()
```

```
swing_state_results.head()
```

```
swing_state_results['Republican'] = swing_state_results['Republican'].astype(int)
```

```
swing_state_results.head()
```

```
swing_states_ed_versus_party_votes = pd.merge(swing_state_ed_facts, swing_state_results, on='area_name')
```

```
swing_states_ed_versus_party_votes.head()
```

In what follows we will be only considering the ratio of high school graduates to party votes (for reasons to be explained), so some columns can be safely dropped.

```
columns_to_be_dropped = ['population', 'high_school_percent', 'higher_ed_percent', 'high_school_grads', 'high_school_dropouts', 'no_higher_school']

swing_states_ed_versus_party_votes = swing_states_ed_versus_party_votes.drop(columns=columns_to_be_dropped)
```

```
swing_states_ed_versus_party_votes.head()
```
