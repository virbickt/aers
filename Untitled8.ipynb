# Statistical Hypothesis Testing

As our metric we will be considering the ratio of those who have bachelor's degree or higher (also to be called ungraduates) to the votes for each party. Let us denote the ratio of *graduates to party votes* as:
 
*   $\hat{p}_{democrats} = \frac{\mathrm{higher\_school\_graduates}}{\mathrm{Democrat}}$ 

(the ratio of the number of those who have a bachelors diploma or higher to the *democrat* votes in that state)
*   $\hat{p}_{repblicans} = \frac{\mathrm{higher\_school\_graduates}}{\mathrm{Republican}}$ 

(the ratio of the number of those who have a bachelors diploma or higher to the *republican* votes in that state)

```
swing_states_ed_versus_party_votes['democrats_to_graduates'] = swing_states_ed_versus_party_votes[['higher_school_graduates', 'Democrat']].apply(
    lambda x: round((x[1] / x[0]), 3),axis=1
)

swing_states_ed_versus_party_votes['republicans_to_graduates'] = swing_states_ed_versus_party_votes[['higher_school_graduates', 'Republican']].apply(
    lambda x: round((x[1] / x[0]), 3),axis=1
)
```

```
swing_states_ed_versus_party_votes.head()
```

Let us denote the observed difference between `democrats_to_graduates` to `republicans_to_graduates` by $\delta_i$ where $i \in \{1, 2, \dotso, 9 \}$

```
swing_states_ed_versus_party_votes['observed_diff'] = swing_states_ed_versus_party_votes['democrats_to_graduates'] - swing_states_ed_versus_party_votes['republicans_to_graduates']
```

```
swing_states_ed_versus_party_votes.head()
```

At this point we have a list of observed differences in our metric among the swing states. Let us further denote the average observed difference by $\delta$:

$ \delta = \frac{\delta_1 + \delta_2 + \dotso + \delta_9}{9}$ where 9 represents the fact the there are 9 swing states. 

 Although, contrary to our assumption, the numbers show that $\hat{p}_{republicans} > \hat{p}_{democrats}$, this does not affect our null hypothesis which we state as follows:

$H_0:$ $\hat{p}_{democrats} - \hat{p}_{republicans} = 0$ 

i. e. there is no significant difference in the ratio of those who have a higher education between two groups (democrat and republican). What this means that that any observed difference is due to random variation.

This is equivalent to:

$\delta = 0$
