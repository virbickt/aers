```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns;sns.set()
```

```
county_facts = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQBakCivIwMI6g1wLMHwvODKVSpY70D-dbeaQanh1i1vl1kMeBvpMApXazTXXa1vPXignSEOAxkGFxM/pub?output=csv')
county_facts_dict = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vSOWGLhyGIf0wGoamZgD91HjkIXoKTjljTesoyPJ_P18y_fnZ66bm24eKw5xM89r1dteTHG8Gpq39zW/pub?output=csv")
primary_results = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTyChEtaOCByzFOBSJSWwSfueA3O5umHQmEpc79qryDn5nnDD5e-ZDM-TmfVPAKLsdEDVn-JHlaTw9a/pub?output=csv')
```

```
state_facts = county_facts[county_facts.isna().any(axis=1)].drop(0).reset_index().drop(columns=['index', 'fips', 'state_abbreviation'])
```

```
state_facts.head()
```

```
swing_states = ['Colorado', 'Florida', 'Iowa', 'Michigan', 'Nevada', 'New Hampshire', 'North Carolina', 'Ohio', 'Pennsylvania', 'Virginia', 'Winconsin']
```

```
swing_state_facts = state_facts.loc[state_facts['area_name'].isin(swing_states)].reset_index().drop(columns=['index'])
```

```
swing_state_facts.head()
```

```
swing_state_ed_facts = swing_state_facts[['area_name', 'PST045214', 'EDU635213', 'EDU685213']].copy().rename(columns={"PST045214": "population", "EDU635213": "high_school_percent", "EDU685213": "higher_ed_percent"})
```

```
swing_state_ed_facts
```

```
swing_state_ed_facts['high_school_grads'] = swing_state_ed_facts[['population', 'high_school_percent']].apply(
    lambda x: (x[0] * (x[1] / 100)), axis=1
)

swing_state_ed_facts['high_school_dropouts'] = swing_state_ed_facts[['population', 'high_school_grads']].apply(
    lambda x: (x[0] - x[1]), axis=1
)

swing_state_ed_facts['higher_school'] = swing_state_ed_facts[['population', 'higher_ed_percent']].apply(
    lambda x: (x[0] * (x[1] / 100)), axis=1
)

swing_state_ed_facts['no_higher_school'] = swing_state_ed_facts[['population', 'higher_school']].apply(
    lambda x: (x[0] - x[1]), axis=1
)
```

```
swing_state_ed_facts
```

```
swing_state_ed_facts.groupby('area_name')[['high_school_grads', 'high_school_dropouts']].sum().T.plot(kind='pie', subplots=True, shadow = True,startangle=90,figsize=(15,10))
```

```
swing_state_ed_facts.groupby('area_name')[['higher_school', 'no_higher_school']].sum().T.plot(kind='pie', subplots=True, shadow = True,startangle=90,figsize=(15,10))
```

```
democrat_swing_results = swing_state_results[swing_state_results['party'] == 'Democrat'].groupby('state')['votes'].sum()
```

```
democrat_swing_results
```

```
republican_swing_results = swing_state_results[swing_state_results['party'] == 'Republican'].groupby('state')['votes'].sum()
```

```
swing_state_results = pd.concat([democrat_swing_results.rename('Democrat'), republican_swing_results.rename('Republican')], axis=1).drop('Colorado')
```

```
swing_state_results # where's Republicans of Colorado?
```

```
swing_state_results = swing_state_results.rename_axis('area_name').reset_index()
```

```
swing_state_results['Democrat_fraction_vote'] = swing_state_results[['Democrat', 'Republican']].apply(
    lambda x: (x[0]/(x[0] + x[1])), axis=1 # it'd be good to round it up to two decimals
)

swing_state_results['Republican_fraction_vote'] = swing_state_results[['Democrat', 'Republican']].apply(
    lambda x: (x[1]/(x[0] + x[1])), axis=1 # it'd be good to round it up to two decimals
)
```
