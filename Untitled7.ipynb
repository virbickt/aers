## Scaling

#### County facts

```
county_columns = county_facts.columns[3:]
```

```
county_features = county_facts[county_columns]
```

```
scaler = StandardScaler()
```

```
scaled_county_features = scaler.fit_transform(county_features)
```

```
fig, axs = plt.subplots(ncols=2, figsize = (22, 9))

sns.boxplot(data=county_features, ax = axs[0])
sns.boxplot(data=scaled_county_features, ax = axs[1])
axs[0].set_title('Before scaling')
axs[1].set_title('After scaling')

axs[0].set_xticklabels(county_columns, rotation=45);
axs[1].set_xticklabels(county_columns, rotation=45);
```

```
data_for_pca = county_facts[county_columns].values
```

```
area_name = county_facts['area_name']
```

```
data_for_pca.shape
```

```
pca_scaled = PCA(2)
data_after_pca = pca_scaled.fit_transform(scaled_county_features)
```

```
pca_unscaled = PCA(2)
data_after_pca_unscaled = pca_unscaled.fit_transform(county_features)
```

```
data_after_pca.shape, data_after_pca_unscaled.shape
```

```
fig, axs = plt.subplots(ncols=2, figsize = (22, 9))

a = sns.scatterplot(x=data_after_pca[:, 0], y=data_after_pca[:, 1], hue=area_name, s = 100, ax = axs[0], legend=None)

b = sns.scatterplot(x=data_after_pca_unscaled[:, 0], y=data_after_pca_unscaled[:, 1], hue=area_name, s = 100, ax = axs[1], legend=None)

a.set_title('Using scaled features')
b.set_title('Using unscaled features')
a.set(xlabel='PC1', ylabel='PC2')
b.set(xlabel='PC1', ylabel='PC2')
#a.legend(bbox_to_anchor = (0.5, -0.30), loc=8, ncol=4);
#b.legend(bbox_to_anchor = (0.5, -0.30), loc=8, ncol=4);
#a.legend.remove()
#b.legend.remove()
```

```
pca_var_2_components = np.round(pca_scaled.explained_variance_ratio_ * 100, 1)
```

```
component_labels = ['PC' + str(x) for x in range(1, len(pca_var_2_components)+1)]
```

```
plt.subplots(figsize = (10, 7))

plt.bar(x=range(1, len(pca_var_2_components)+1), height=pca_var_2_components, tick_label=component_labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.show
```
